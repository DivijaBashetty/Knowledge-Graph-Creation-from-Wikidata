```python
k=0
cols=['pLabel.value','q.value','rLabel.value']
df=pd.DataFrame(columns=cols)
for i in qlist:
  query=""" select  ?pLabel ?propLabel ?rLabel   where  {  BIND (wd:""" + str(i) + """ AS ?r) ?p ?q ?r. ?prop wikibase:directClaim ?q FILTER contains(str(?p),'Q') SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE]". }		}"""
  k=k+1
  #print("\n"+query+"\n")
  sparql.setQuery(query)
  sparql.setReturnFormat(JSON)
  result = sparql.query().convert()
  results_df=pd.json_normalize(result['results']['bindings'])
  if(results_df.shape[1]!=0):
    '''for s in results_df['q.value']:
      j=str(s)
      x=j.split('/')
      results_df['qLabel.value']=x[-1]'''
    results_df=results_df[['pLabel.value','propLabel.value','rLabel.value']]
    print(k)
    print(results_df[['pLabel.value','propLabel.value','rLabel.value']])
    print("\n--------------------------------------------------------------------------------------------\n")
    print(df.shape,results_df.shape)
    results_df=results_df.drop_duplicates()
    df=pd.concat([df,results_df])
    print(df.shape)
    df=df.reset_index(drop=True)
    #print(df.shape)
```

```python
for i in range(0,df['pLabel.value'].shape[0]):
  if(i%250==0):
    print(i)
  x=df['pLabel.value'][i]
  y=df['propLabel.value'][i]
  z=df['rLabel.value'][i]
  l=get_entity_dict_from_api(x)
  label=WikidataItem(l)
  df['pLabel.value'][i]=label.get_label()
  l=get_entity_dict_from_api(y)
  label=WikidataProperty(l)
  df['propLabel.value'][i]=label.get_label()
  l=get_entity_dict_from_api(z)
  label=WikidataItem(l)
  df['rLabel.value'][i]=label.get_label()
```
